# Gradient Descent

Consider the bivariate function f : $R^2 → R$ that is defined as follows:
$f(w) = (w_1^2+w_2-11)^2+(w_1+w_2^2-7)^2$     
$df/dw_1 = 4w_1(w_1^2+w_2-11)+2(w_1+w_2^2-7)$
$df/dw_2 = 2(w_1^2+w_2-11)+4w_2(w_1+w_2^2-7)$

### 1. Initialize gradient descent from $w_0 = [0,-4]^T$

#### Step size = 0.005

```python
import numpy as np
import numpy.linalg as la
def f(w1,w2):
    return (w1**2+w2-11)**2 + (w1+w2**2-7)**2
def dw1(w):
    w1 = w[0]; w2 = w[1];
    return 4*w1*(w1**2+w2-11)+2*(w1+w2**2-7)
def dw2(w):
    w1 = w[0]; w2 = w[1];
    return 2*(w1**2+w2-11)+4*w2*(w1+w2**2-7)
def grad(w):
    g = np.zeros(shape=(2,1))
    g[0]=dw1(w)
    g[1]=dw2(w)
    return g
def new_w(w,l):
    w1 = w[0]; w2 = w[1];
    w1_new =  w1 - l*dw1(w)
    #print(w1_new)
    w2_new = w2 - l*dw2(w)
    #print(w2_new)
    return np.reshape([w1_new,w2_new],(2,1))
w0 = np.reshape([0,-4],(2,1))
print("w0 is: \n",w0)
it = 0
l = 0.005
while(1):
    it = it + 1
    #print("Iteration ",it,":")
    wprev = w0
    w0 = new_w(w0,l)
    #print("f(w) = ",f(w0[0],w0[1]))
    if(la.norm(grad(w0))**2<10**(-12)):
        print("Iterations: ",it)
        break
    if (abs(w0[0]-wprev[0])==0 and abs(w0[1]-wprev[1])==0):
        print("Iterations: ",it)
        break
        
print("w is: \n",w0)
```

#### Step size = 0.01

```python
w0 = np.reshape([0,-4],(2,1))
print("w0 is: \n",w0)
it = 0
l = 0.01
while(1):
    it = it + 1
    #print("Iteration ",it,":")
    wprev = w0
    w0 = new_w(w0,l)
    #print("f(w) = ",f(w0[0],w0[1]))
    if(la.norm(grad(w0))**2<10**(-12)):
        print("Iterations: ",it)
        break
    if (abs(w0[0]-wprev[0])==0 and abs(w0[1]-wprev[1])==0):
        print("Iterations: ",it)
        break
        
print("w is: \n",w0)
```

The number of iterations is inversely proportional to the step size: The smaller the step size the more iterations needed. We can say that the relationship is linear since a step size of 0.005 required approximately double the iterations and required by a step size of 0.01, which is double of 0.005.

# Parameter Estimation

```python
import numpy as np
import numpy.linalg as la
from scipy.stats import multivariate_normal
m = [-1,0,4,1,0.5]
m = np.asarray(m)
C = 2*np.identity(5)

j = np.arange(1,7)
Nr = np.zeros(6)
eN = np.zeros(6)
print(Nr.shape)
for i in j:
    N = 10**i
    Nr[i-1]=N
    x = np.random.uniform(size=(5, 5))
    #print(x)
    X = multivariate_normal(m,C,N)
    sum = np.zeros(shape=(5,))
    for n in range(1,N+1):
        xn = X.rvs()
        sum = sum + xn
    mu_n = sum/N
    print(f"\nj = {i}: mu_N is: \n",mu_n)
    eN[i-1] = la.norm(mu_n-m,2)**2
    print(f"The instantaneous error for N = 10^{i} is: ",eN[i-1])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111)
plt.yscale('log')
plt.xscale('log')
ax.plot(Nr, eN, c='c')
ax.scatter(Nr,eN,c='c',marker='o')
ax.set_title("Error vs N")
ax.set_xlabel('N (log)')
ax.set_ylabel('error_N (log)')
```

The relationship between data samples N and error is inversely proportional. The higher the number of data samples, the smaller the error.
