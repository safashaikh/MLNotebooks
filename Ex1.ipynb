### 1. Plot GDP as function of the year

```python
year = [1930+10*i for i in range(9)]
gdp = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
print(year); print(gdp)
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

plt.plot(year, gdp, marker='o', linestyle='dashed')
plt.show()
```

A possible mathematical relationship (by looking at the shape of the curve) can be some quadratic form such as $y=ax^2+bx+c$. Let's guess $y=x^2+1$.

### 3
Let's plot this prediction:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
year = [1930+10*i for i in range(9)]
gdp = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
x = [i - 1930 for i in year]
print(x)
plt.plot(year, gdp, marker='o', linestyle='dashed')
plt.plot(year, [(i**2)+1 for i in x], marker='o', linestyle='dashed', color='green')
plt.show()
```

This function increases too quickly; we can try widening the curve by decreasing the $a$ value:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
year = [1930+10*i for i in range(9)]
gdp = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
x = [i - 1930 for i in year]
plt.plot(year, gdp, marker='o', linestyle='dashed')
plt.plot(year, [(i**2)/400 + 1 for i in x], marker='o', linestyle='dashed', color='green')
print([(i**2)/400 + 1 for i in x])
print(gdp)
plt.show()
```

It seems like this function is a good fit. The $a$ coefficient of $1/400$ was obtained using a guess and check method. A more mathematically sound estimation can be obtained using least squares fit or singular value decomposition.
